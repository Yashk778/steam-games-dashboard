```python
import requests

page = 1

while True:
    url = f"https://healthrevu.com/providers/hospital?page={page}"
    response = requests.get(url)

    if not response.text.strip():
        break

    print("Page", page, "exists")
    page += 1
```

```python
import requests
from bs4 import BeautifulSoup

data = []

for page in range(1,101):

    webpage = requests.get(f'https://healthrevu.com/providers/hospital?page={page}')

    soup = BeautifulSoup(webpage.text, 'lxml')

    cards = soup.find_all('div', class_='provider-item')

    for card in cards:

        name = card.find('strong').text.strip()

        location = card.find('address').find('small').text.strip()

        rating_tag = card.find('div', class_='col-12 col-md-4 ms-md-3')

        if rating_tag:
            rating_block = rating_tag.text.strip().split()

            if len(rating_block) >= 4:
                rating = rating_block[0]
                reviews = rating_block[-2]
            else:
                rating = None
                reviews = None
        else:
            rating = None
            reviews = None

        data.append({
            'hospital_name': name,
            'location': location,
            'rating': rating,
            'reviews': reviews
        })

    print("scraped_page:", page)
```

```python
import pandas as pd

df = pd.DataFrame(data)

df.info()
```
